```python
def count_freq(df,colname):
    dic={}
    for ele in df[colname]:
        if ele in dic:
            dic[ele]+=1
        else:
            dic[ele]=1
    return dic
```

```python
def count_freq_lst(df, name):
    #solve decoding Chinese problem
    
    genreeng={
     '剧情': 'Drama',
     '冒险': 'Action',
     '动作': 'Action',
     '喜剧': 'Comic',
     '动画': 'Cartoon',
     '奇幻': 'Fiction',
     '恐怖': 'Horror',
     '悬疑': 'Horror',
     '惊悚': 'Horror',
     '战争': 'War',
     '爱情': 'Romance',
     '犯罪': 'Crime',
     '科幻': 'Fiction',
     '纪录片': 'Documentary'
    }
    countryeng={
     '美国': 'US',
     '英国': 'UK',
     '日本': 'Jap',
     '中国': 'Mainland',
     '中国大陆': 'Mainland',
     '香港': 'HK',
     '韩国': 'Korea',
     '加拿大': 'Canada',
     '法国': 'France',
     '俄罗斯': 'Russia',
     '印度': 'India',
     '德国': 'German',
     '丹麦': 'Denmark',
     '泰国': 'Thai',
     '荷兰': 'Nether',
     '台湾': 'TaiWan',
     '加拿大C': 'Canada',
     '意大利': 'Italia'
        
    }
    
    
    if name=="Genre":
        engname=genreeng
    elif name=="District":
        engname=countryeng
        
    dic={}
    dic['Else']=0
    for ele in df[name]:
        if ele[0] in engname:
            if engname[ele[0]] in dic:
                dic[engname[ele[0]]]+=1
            else:
                dic[engname[ele[0]]]=1
        else:
            dic['Else']+=1
    return dic
```

```python
countgenre=count_freq_lst(df,"Genre")
commongenre=[key for key in countgenre.keys() if countgenre[key]>=2]
comcount=[countgenre[ele] for ele in commongenre]
sns.barplot(x=commongenre, y=comcount)
```

```python
def name_score(df, name):
    #solve decoding Chinese problem
    
    genreeng={
     '剧情': 'Drama',
     '冒险': 'Action',
     '动作': 'Action',
     '喜剧': 'Comic',
     '动画': 'Cartoon',
     '奇幻': 'Fiction',
     '恐怖': 'Horror',
     '悬疑': 'Horror',
     '惊悚': 'Horror',
     '战争': 'War',
     '爱情': 'Romance',
     '犯罪': 'Crime',
     '科幻': 'Fiction',
     '纪录片': 'Documentary'
    }
    countryeng={
     '美国': 'US',
     '英国': 'UK',
     '日本': 'Jap',
     '中国': 'Mainland',
     '中国大陆': 'Mainland',
     '香港': 'HK',
     '韩国': 'Korea',
     '加拿大': 'Canada',
     '法国': 'France',
     '俄罗斯': 'Russia',
     '印度': 'India',
     '德国': 'German',
     '丹麦': 'Denmark',
     '泰国': 'Thai',
     '荷兰': 'Nether',
     '台湾': 'TaiWan',
     '加拿大C': 'Canada',
     '意大利': 'Italia'
        
    }
    
    
    if name=="Genre":
        engname=genreeng
    elif name=="District":
        engname=countryeng
        
    dic_tot={}
    dic_tot['Else']=0
    rate=df['Rating'].tolist()
    for i in range(df.shape[0]):
        if df[name][i][0] in engname:
            if engname[df[name][i][0]] in dic_tot:
                dic_tot[engname[df[name][i][0]]]+=float(rate[i])
            else:
                dic_tot[engname[df[name][i][0]]]=float(rate[i])
        else:
            dic_tot['Else']+=float(rate[i])
            
    dic_freq=count_freq_lst(df,name)
    
    dic_new={}
    for key in dic_tot.keys():
        if dic_freq[key]==0:
            continue
        else:
            dic_new[key]=float(dic_tot[key])/float(dic_freq[key])
    
    return dic_new
    
```

```python
name_score(df, 'Genre')
```

```python
dic_type_score=name_score(df, 'Genre')
genre=[key for key in dic_type_score.keys() if count_freq_lst(df,'Genre')[key]>5]
avg=[dic_type_score[ele] for ele in genre]
sns.barplot(x=genre, y=avg)
```

```python
dic_type_score=name_score(df, 'District')
district=[key for key in dic_type_score.keys() if count_freq_lst(df,'District')[key]>20]
avg=[dic_type_score[ele] for ele in district]
sns.barplot(x=district, y=avg)
```
